```python
import dlib
import cv2
import numpy as np
import sys
import re
import os
from os.path import dirname, join, abspath
from PIL import Image
import time
import face_recognition
import matplotlib.pyplot as plt
```

```python
class Training:
    def __init__(self):
        self.base_path = os.path.dirname(os.path.realpath('__file__'))
        self.train_path = join(self.base_path, 'train')
        self.list_image_path = join(self.base_path, 'list_images')
        self.img_train = os.listdir(self.train_path)
        self.models_path = join(self.base_path, 'models')
        self.weight_path = join(self.models_path, 'best.npy')
        self.class_name_path = join(self.models_path, 'class_name.npy')

            
    def find_class_name(self, img_train):
        images = []
        class_name = []
        no_ind = self.get_data_users()
        for img in img_train:
            cur_img = cv2.imread(f'{self.train_path}/{img}')
            user = os.path.splitext(img)[0].upper()
            res = re.findall(r'\w+', user)
            if res[0] not in no_ind:
                imgs = Image.open(f'{self.train_path}/{img}')
                imgs.save(f'{self.list_image_path}/{img.upper()}')
                images.append(cur_img)
                class_name.append(os.path.splitext(img)[0].upper())
        return images, class_name

    def find_encode(self, images):
        encode_list = []
        for img in images:
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            encode = face_recognition.face_encodings(img)[0]
            encode_list.append(encode.tolist())
        return encode_list
                    
    def append_params(self, encode_list, class_name):
        try:
            weight = np.load(self.weight_path)
            list_weight = weight.tolist()
            list_weight.append(encode_list[0])
            arr = np.array(list_weight)
            data_class_name = np.load(self.class_name_path)
            list_data_class_name = data_class_name.tolist()
            list_data_class_name.append(class_name[0])
            np.save(self.weight_path, arr)
            np.save(self.class_name_path, list_data_class_name)
        except:
            pass
#             np.save(self.weight_path, encode_list)
#             np.save(self.class_name_path, class_name)
            
    def get_data_users(self):
        img_list = os.listdir(self.list_image_path)
        list_noind = []
        for user in img_list:
            usr = os.path.splitext(user)[0]
            res = re.findall(r'\w+', usr)
            list_noind.append(res[0])
        arr_users = np.array(list_noind)
        return arr_users
```

```python
train = Training()
img, class_name = train.find_class_name(train.img_train)
encode_list = train.find_encode(img)
if encode_list != []:
    train.append_params(encode_list, class_name)
else:
    print('No execute task')
```
