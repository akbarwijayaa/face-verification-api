```python
import dlib
import cv2
import numpy as np
import sys
import os
from os.path import dirname, join, abspath
import face_recognition
import matplotlib.pyplot as plt
```

```python
base_path = os.path.dirname(os.path.realpath('__file__'))
img_path = join(base_path, 'img')
img_train = os.listdir(img_path)
```

```python
images = []
class_name = []
for img in img_train:
    cur_img = cv2.imread(f'{img_path}/{img}')
    images.append(cur_img)
    class_name.append(os.path.splitext(img)[0])
```

```python
def findEncodings(images):
    encode_list = []
    for img in images:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        encode = face_recognition.face_encodings(img)[0]
        encode_list.append(encode)
    return encode_list
```

```python
import cv2

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    cv2.imshow('frame', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
img_train = face_recognition.load_image_file('img/me_train.jpg')
img_train = cv2.cvtColor(img_train, cv2.COLOR_BGR2RGB)
```

```python
img_test = face_recognition.load_image_file('img/akbar.jpg')
img_test = cv2.cvtColor(img_test, cv2.COLOR_BGR2RGB)
```

```python
faceLoc = face_recognition.face_locations(img_train)[0]
encode_train = face_recognition.face_encodings(img_train)[0]
train = cv2.rectangle(img_train, (faceLoc[3], faceLoc[0]), (faceLoc[1], faceLoc[2]), (255, 0, 0), 2)
plt.imshow(train)
```

```python
faceLoc = face_recognition.face_locations(img_test)[0]
encode_test = face_recognition.face_encodings(img_test)[0]
test = cv2.rectangle(img_test, (faceLoc[3], faceLoc[0]), (faceLoc[1], faceLoc[2]), (255, 0, 0), 2)
plt.imshow(test)
```

```python
result = face_recognition.compare_faces([encode_train], encode_test)
distance = face_recognition.face_distance([encode_train], encode_test)
print(result, distance)
```
